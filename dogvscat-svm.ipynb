```python
import os
import cv2
import numpy as np
from skimage.feature import hog

images = []
labels = []

train_cat_dir = "/kaggle/input/cats-vs-dogs-1000/dogs_cats_sample_1000/train/cats"
train_dog_dir = "/kaggle/input/cats-vs-dogs-1000/dogs_cats_sample_1000/train/dogs"

# Load CAT images
for file in os.listdir(train_cat_dir):
    if file.lower().endswith(".jpg"):
        img_path = os.path.join(train_cat_dir, file)
        img = cv2.imread(img_path)
        img = cv2.resize(img, (128, 128))
        img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        hog_features = hog(img_gray,
                           orientations=9,
                           pixels_per_cell=(8, 8),
                           cells_per_block=(2, 2),
                           block_norm='L2-Hys')

        images.append(hog_features)
        labels.append(0)   # cat

# Load DOG images
for file in os.listdir(train_dog_dir):
    if file.lower().endswith(".jpg"):
        img_path = os.path.join(train_dog_dir, file)
        img = cv2.imread(img_path)
        img = cv2.resize(img, (128, 128))
        img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        hog_features = hog(img_gray,
                           orientations=9,
                           pixels_per_cell=(8, 8),
                           cells_per_block=(2, 2),
                           block_norm='L2-Hys')

        images.append(hog_features)
        labels.append(1)   # dog

images = np.array(images)
labels = np.array(labels)

print("Images:", len(images))
print("Labels:", len(labels))
print("One feature vector shape:", images[0].shape)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    images, labels, test_size=0.2, random_state=42
)
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

svm = LinearSVC()
svm.fit(X_train, y_train)

pred = svm.predict(X_test)
print("Accuracy:", accuracy_score(y_test, pred))
```
